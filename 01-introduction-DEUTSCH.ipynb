So what's Haskell?
------------------

Haskell ist eine __reine
funktionale Programmiersprache__. In imperativen Sprachen erreicht man Dinge
indem man dem Computer eine Folge von Aufgaben gibt, die er dann ausführt
sie aus. Während er sie ausführt, kann er seinen Zustand ändern. Zum Beispiel setzt man
Variable `a` auf 5 und führt dann etwas aus und setzt sie dann auf etwas
etwas anderes. Man hat Kontrollflussstrukturen, um eine Aktion mehrmals auszuführen.
In der rein funktionalen Programmierung sagt man dem Computer nicht, was er
tun soll, sondern man sagt ihm, was etwas *ist*.
Die Fakultät einer Zahl ist das Produkt aller Zahlen von 1 bis zu dieser Zahl, die Summe
einer Liste von Zahlen ist die erste Zahl plus die Summe aller anderen
Zahlen, und so weiter. Man drückt das in Form von Funktionen aus. Man kann auch
nicht eine Variable  auf einen bestimmten Wert setzen und sie später auf einen anderen. Wenn du sagst, dass a 5 ist, kannst du später nicht sagen, dass es etwas anderes ist, weil du gerade gesagt hast, es sei 5.
weil du gerade gesagt hast, es sei 5. Bist du etwa ein Lügner? Also in
rein funktionalen Sprachen hat eine Funktion keine Nebeneffekte. Das einzige
was eine Funktion tun kann, ist etwas zu berechnen und es als
Ergebnis zurück zu geben. Auf den ersten Blick scheint das etwas einschränkend zu sein, aber es hat tatsächlich einige
nette Konsequenzen: Wenn eine Funktion zweimal mit denselben Parametern aufgerufen wird
Parametern aufgerufen wird, gibt sie garantiert das gleiche Ergebnis zurück. Das nennt man
Referenztransparenz genannt, und sie erlaubt dem Compiler nicht nur
über das Verhalten des Programms zu schließen, sondern es erlaubt Ihnen auch, leicht
ableiten (und sogar beweisen), dass eine Funktion korrekt ist, und dann
komplexere Funktionen durch Zusammenkleben einfacher Funktionen zu erstellen.

<img src="img/lazy.png" title="lazy" style="float:right;margin-left:2em;" /> Haskell ist __lazy__. Das
bedeutet, dass Haskell, wenn es nicht ausdrücklich anders gesagt wird, keine
Funktionen ausführt und Dinge berechnet, bis es wirklich gezwungen ist, Ihnen ein
Ergebnis zu zeigen. Das passt gut zur referenziellen Transparenz und erlaubt es Ihnen
Programme als eine Reihe von __Transformationen von Daten__ zu betrachten. Es erlaubt auch
erlaubt coole Dinge wie unendliche Datenstrukturen. Nehmen wir an, Sie haben eine
unveränderliche Liste von Zahlen `xs = [1,2,3,4,5,6,7,8]` und eine Funktion `doubleMe`
die jedes Element mit 2 multipliziert und dann eine neue Liste zurückgibt. Wenn wir
unsere Liste in einer imperativen Sprache mit 8 multiplizieren wollen und dazu
`doubleMe(doubleMe(doubleMe(xs)))`, würde die Liste wahrscheinlich einmal durchlaufen
Liste einmal durchlaufen, eine Kopie erstellen und diese dann zurückgeben. Dann würde es die Liste
die Liste noch zwei weitere Male durchlaufen und das Ergebnis zurückgeben. In einer lazy language,
Aufrufen von `doubleMe` für eine Liste, ohne dass das Ergebnis angezeigt wird
endet damit, dass das Programm Ihnen sagt: "Ja, ja, ich mache es
später!". Aber sobald man das Ergebnis sehen will, sagt das erste `doubleMe` dem zweiten
dem zweiten, dass es das Ergebnis haben will, und zwar jetzt! Der zweite sagt das dem
dritten und der dritte gibt widerwillig eine verdoppelte 1 zurück, also eine 2.
Der zweite erhält das Ergebnis und gibt dem ersten eine 4 zurück. Der
erste sieht das und sagt dir, dass das erste Element 8 ist. Sie macht also nur
nur einmal durch die Liste und auch nur dann, wenn man es wirklich braucht. Das
Wenn man etwas von einer faulen Sprache will, kann man einfach ein paar
Daten nehmen und sie effizient umwandeln und ändern, so dass sie dem ähneln, was
was Sie am Ende wollen.

Haskell ist __elegant und prägnant__. Weil es eine Menge von High-Level
Konzepte verwendet, sind Haskell-Programme normalerweise kürzer als ihre imperativen
Äquivalente. Und kürzere Programme sind leichter zu warten als längere
und haben weniger Fehler.

Haskell wurde von einigen __wirklich klugen Leuten__ (mit Doktortitel) entwickelt. Die Arbeit an
Haskell begann 1987, als ein Komitee von Forschern zusammenkam, um
eine tolle Sprache zu entwickeln. Im Jahr 2003 wurde der Haskell-Report veröffentlicht,
der eine stabile Version der Sprache definiert.
